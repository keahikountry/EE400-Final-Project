## Face Detection on the Received PNG
Now we will use OpenCV to detect faces in the `output.png` image and display the results.

```python
import numpy as np
import matplotlib.pyplot as plt


# --- RGB565 -> bytes (big-endian: high byte, low byte) ---
raw_bytes = np.empty(len(hex_words) * 2, dtype=np.uint8)
for i, word in enumerate(hex_words):
    val = word  # already an int like 0xABCD
    raw_bytes[2*i]   = (val >> 8) & 0xFF
    raw_bytes[2*i+1] =  val       & 0xFF

# --- Convert RGB565 to RGB888 ---
def rgb565_to_rgb888(data: np.ndarray) -> np.ndarray:
    vals = (data[0::2].astype(np.uint16) << 8) | data[1::2].astype(np.uint16)
    r = ((vals >> 11) & 0x1F) << 3
    g = ((vals >> 5)  & 0x3F) << 2
    b = ( vals        & 0x1F) << 3
    return np.stack((r, g, b), axis=1).astype(np.uint8)

pixels = rgb565_to_rgb888(raw_bytes)

# --- Pick image dimensions ---
total_pixels = pixels.shape[0]

# Try common microdisplay sizes first; falls back to any factor pair
preferred = [(72,96), (96,72), (120,160), (160,120), (128,128),
             (240,240), (240,135), (135,240), (320,240), (240,320),
             (160,80), (80,160), (96,160), (160,96)]
hw = next(((h,w) for h,w in preferred if h*w == total_pixels), None)

if hw is None:
    # fallback: pick the most square factor pair
    factors = [(h, total_pixels//h) for h in range(1, int(total_pixels**0.5)+1) if total_pixels % h == 0]
    hw = min(factors, key=lambda hw: abs(hw[0]-hw[1]))  # closest to square

height, width = hw

# --- Reshape and display ---
image = pixels.reshape(height, width, 3)

plt.imshow(image)
plt.axis("off")
plt.savefig("output.png", bbox_inches="tight", pad_inches=0)
plt.show()

```

# Face Detection Model Compression Workflow (TensorFlow)
This notebook will:
1. Train a simple face detection classifier using TensorFlow/Keras.
2. Apply model compression techniques:
   - Knowledge Distillation (KD)
   - Pruning
   - Quantization-Aware Training (QAT)
   - Post-Training Quantization (PTQ)
3. Compare model size, accuracy, and latency for each configuration.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers
import tensorflow_model_optimization as tfmot
from sklearn.model_selection import train_test_split
print('TensorFlow version:', tf.__version__)
```

## Dataset Preparation
For demonstration, we will use the LFW (Labeled Faces in the Wild) dataset from scikit-learn. In practice, you should use your own dataset or a more suitable face detection dataset.

```python
from sklearn.datasets import fetch_lfw_people
from sklearn.utils import shuffle
 
# Load LFW people dataset (faces)
lfw = fetch_lfw_people(min_faces_per_person=20, resize=0.4)
X, y = lfw.images, lfw.target
 
# For demonstration, treat all images as 'face' (label 1) and generate random 'no-face' images (label 0)
X_face = X
y_face = np.ones(len(X_face))
 
# Generate random noise images as 'no-face'
X_noface = np.random.rand(len(X_face), X_face.shape[1], X_face.shape[2]) * 255
y_noface = np.zeros(len(X_noface))
 
# Combine and shuffle
X_all = np.concatenate([X_face, X_noface], axis=0)
y_all = np.concatenate([y_face, y_noface], axis=0)
X_all, y_all = shuffle(X_all, y_all, random_state=42)
 
# Normalize images
X_all = X_all.astype('float32') / 255.0
X_all = np.expand_dims(X_all, -1)  # Add channel dimension
 
# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_all, y_all, test_size=0.2, random_state=42)
print('Train shape:', X_train.shape, 'Test shape:', X_test.shape)
```

## Baseline Model Training
We will train a simple CNN classifier to distinguish between face and no-face images.

```python
def build_baseline_model(input_shape):
    model = keras.Sequential([
        layers.Conv2D(16, (3, 3), activation='relu', input_shape=input_shape),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(32, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

baseline_model = build_baseline_model(X_train.shape[1:])
baseline_model.summary()

history = baseline_model.fit(X_train, y_train, epochs=5, batch_size=32, validation_split=0.1)
```

```python
baseline_loss, baseline_acc = baseline_model.evaluate(X_test, y_test, verbose=2)
print(f'Baseline Model Test Accuracy: {baseline_acc:.4f}')
```

## Knowledge Distillation (KD)
We will train a smaller student model using the baseline model as a teacher.

```python
def build_student_model(input_shape):
    model = keras.Sequential([
        layers.Conv2D(8, (3, 3), activation='relu', input_shape=input_shape),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(8, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

student_model = build_student_model(X_train.shape[1:])
student_model.summary()

# Knowledge distillation training loop
teacher_preds = baseline_model.predict(X_train)
alpha = 0.5  # weight for distillation loss
temperature = 3.0
 
def distillation_loss(y_true, y_pred):
    y_true = tf.cast(y_true, tf.float32)
    hard_loss = keras.losses.binary_crossentropy(y_true, y_pred)
    soft_loss = keras.losses.binary_crossentropy(tf.nn.sigmoid(teacher_preds/temperature), tf.nn.sigmoid(y_pred/temperature))
    return alpha * hard_loss + (1 - alpha) * soft_loss

student_model.fit(X_train, y_train, epochs=5, batch_size=32, validation_split=0.1)
```

```python
student_loss, student_acc = student_model.evaluate(X_test, y_test, verbose=2)
print(f'Student Model Test Accuracy: {student_acc:.4f}')
```

## Model Pruning
We will prune the baseline model to reduce its size while maintaining accuracy.

```python
prune_low_magnitude = tfmot.sparsity.keras.prune_low_magnitude
pruning_params = {
    'pruning_schedule': tfmot.sparsity.keras.PolynomialDecay(initial_sparsity=0.0, final_sparsity=0.5, begin_step=0, end_step=1000)
}

pruned_model = prune_low_magnitude(build_baseline_model(X_train.shape[1:]), **pruning_params)
pruned_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

callbacks = [tfmot.sparsity.keras.UpdatePruningStep()]
pruned_model.fit(X_train, y_train, epochs=5, batch_size=32, validation_split=0.1, callbacks=callbacks)
```

```python
pruned_model = tfmot.sparsity.keras.strip_pruning(pruned_model)
pruned_loss, pruned_acc = pruned_model.evaluate(X_test, y_test, verbose=2)
print(f'Pruned Model Test Accuracy: {pruned_acc:.4f}')
```

## Quantization-Aware Training (QAT)
We will apply quantization-aware training to further compress the model.

```python
qat_model = tfmot.quantization.keras.quantize_model(build_baseline_model(X_train.shape[1:]))
qat_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
qat_model.fit(X_train, y_train, epochs=5, batch_size=32, validation_split=0.1)
```

```python
qat_loss, qat_acc = qat_model.evaluate(X_test, y_test, verbose=2)
print(f'QAT Model Test Accuracy: {qat_acc:.4f}')
```

## Post-Training Quantization (PTQ)
We will apply post-training quantization to the baseline model and evaluate its performance.

```python
import tempfile
converter = tf.lite.TFLiteConverter.from_keras_model(baseline_model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_quant_model = converter.convert()

# Save the quantized model to a file
with tempfile.NamedTemporaryFile(delete=False, suffix='.tflite') as f:
    f.write(tflite_quant_model)
    quant_model_path = f.name
print('Quantized model saved to:', quant_model_path)
```

## Model Comparison and Reporting
We will now compare the size, accuracy, and latency of all models: baseline, student (KD), pruned, QAT, and PTQ.

```python
import os
import time

def get_model_size(model):
    _, fname = tempfile.mkstemp('.h5')
    model.save(fname, include_optimizer=False)
    size = os.path.getsize(fname) / 1024
    os.remove(fname)
    return size

def get_latency(model, X):
    start = time.time()
    model.predict(X[:32])
    return (time.time() - start) / 32

results = {
    'Model': [],
    'Size (KB)': [],
    'Accuracy': [],
    'Latency (s/sample)': []
}

# Baseline
results['Model'].append('Baseline')
results['Size (KB)'].append(get_model_size(baseline_model))
results['Accuracy'].append(baseline_acc)
results['Latency (s/sample)'].append(get_latency(baseline_model, X_test))

# Student
results['Model'].append('Student (KD)')
results['Size (KB)'].append(get_model_size(student_model))
results['Accuracy'].append(student_acc)
results['Latency (s/sample)'].append(get_latency(student_model, X_test))

# Pruned
results['Model'].append('Pruned')
results['Size (KB)'].append(get_model_size(pruned_model))
results['Accuracy'].append(pruned_acc)
results['Latency (s/sample)'].append(get_latency(pruned_model, X_test))

# QAT
results['Model'].append('QAT')
results['Size (KB)'].append(get_model_size(qat_model))
results['Accuracy'].append(qat_acc)
results['Latency (s/sample)'].append(get_latency(qat_model, X_test))

# PTQ (TFLite)
ptq_size = os.path.getsize(quant_model_path) / 1024
results['Model'].append('PTQ (TFLite)')
results['Size (KB)'].append(ptq_size)
results['Accuracy'].append('N/A')
results['Latency (s/sample)'].append('N/A')

import pandas as pd
df = pd.DataFrame(results)
display(df)
```
